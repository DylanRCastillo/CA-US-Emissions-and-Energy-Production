```python
import pandas as pd
```

```python
plants = 'plants.csv'
plants_2010 = 'List_of_plants_for_all_fuels_United_States_all_sectors_2010.csv'
plants_df = pd.read_csv(plants)
plants2010_df = pd.read_csv(plants_2010)
plants2011_df = pd.read_csv('List_of_plants_for_all_fuels_United_States_all_sectors2011.csv')
```

```python
plants_dfs = plants_df.loc[plants_df['year']==2010]
plants_dfs = plants_dfs.rename(columns={'plant_id':'Plant Code','plant_name_cap':'Plant Name'})
plants_dfs.head()
```

```python
plants2010_df = plants2010_df.loc[plants2010_df['State'] == 'CA']
```

```python
merged_df = pd.merge(plants_dfs,plants2010_df,on=['Plant Code','Plant Name'])
```
