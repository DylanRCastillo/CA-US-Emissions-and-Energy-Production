# Dependencies and Set up

```python
# Dependencies and SetUp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import scipy.stats as stats
import os
```

```python
# Load CSV file and display into a DataFrame
R_emission_csv = os.path.join('Resources','use_tot_realgdp.csv')
R_emissions_df = pd.read_csv(R_emission_csv,skiprows=2)
R_emissions_df.head()
```

```python
# Clean DataFrame to only show years from 2010 to 2019
R_data_df = R_emissions_df.drop(R_emissions_df.columns[1:51],axis=1)
R_clean_df = R_data_df.drop(R_data_df.columns[11:],axis=1)
R_clean_df.head()
```

```python
# Transpose dataframe, rename columns with first row, remove fisrt row and change the first column name to Year
R_data_clean_df = R_clean_df.T
R_data_clean_df = R_data_clean_df.reset_index()
R_data_clean_df.columns = R_data_clean_df.iloc[0]
R_data_clean_df = R_data_clean_df.drop(R_data_clean_df.index[[0]])
R_data_clean_df = R_data_clean_df.rename(columns={'State':'Year'})

# Convert data to numeric values
R_states_usage_int = R_data_clean_df.apply(pd.to_numeric)
R_states_usage_int
```

```python
# Store clean data to a csv file
R_states_usage_int.to_csv('Resources/Total Energy Consumption 2010 to 2019.csv',index=False)
```
