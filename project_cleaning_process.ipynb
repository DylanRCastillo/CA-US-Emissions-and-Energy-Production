```python
# Dependencies and SetUp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import scipy.stats as stats
import os
```

```python
# Load CSV file and display into a DataFrame
R_emission_csv = os.path.join('Resources','use_tot_realgdp.csv')
R_emissions_df = pd.read_csv(R_emission_csv,skiprows=2)
R_emissions_df.head()
```

```python
# Clean DataFrame to only show years from 2010 to 2019
R_data_df = R_emissions_df.drop(R_emissions_df.columns[1:51],axis=1)
R_clean_df = R_data_df.drop(R_data_df.columns[11:],axis=1)
R_clean_df.head()
```

```python
# Transpose dataframe, rename columns with first row, remove fisrt row and change the first column name to Year
R_data_clean_df = R_clean_df.T
R_data_clean_df = R_data_clean_df.reset_index()
R_data_clean_df.columns = R_data_clean_df.iloc[0]
R_data_clean_df = R_data_clean_df.drop(R_data_clean_df.index[[0]])
R_data_clean_df = R_data_clean_df.rename(columns={'State':'Year'})

# Convert data to numeric values
R_states_usage_int = R_data_clean_df.apply(pd.to_numeric)
R_states_usage_int
```

```python
# Store clean data to a csv file
R_states_usage_int.to_csv('Resources/Total Energy Consumption 2010 to 2019.csv',index=False)
```

```python
#Clean dataframe
#New dataframe remove units & source key
df_data_dylan = data_dylan.copy()
to_drop_dylan = ['units', 'source key']
df_data_dylan.drop(to_drop_dylan, inplace=True, axis=1)
```

```python
#Show new dataframe
df_data_dylan = df_data_dylan.reset_index()
df_data_dylan.drop(['index'], inplace=True, axis=1)
df_data_dylan
```

```python
#Massage the data using pandas.melt()
#description is id_vars and each month is value_vars
#pd.melt(df_data, id_vars=['description'], value_vars=[''])
columns_name_dylan = df_data_dylan.columns.values.tolist()
columns_name_dylan.remove(columns_name_dylan[0])
value_vars_dylan=columns_name_dylan
id_vars_dylan='description'

data_dylan = pd.melt(df_data_dylan, id_vars=['description'], value_vars=value_vars_dylan).drop(columns=['variable'])
data_dylan = data_dylan.replace('--', 0)
data_dylan = data_dylan.replace('NM', 0)
data_dylan["value"] = pd.to_numeric(data_dylan["value"], downcast="float")
data_dylan = data_dylan.rename(columns={'description':'United States Power Source', 'value': 'Energy Production (Thousand Megawatthours)'})
```

```python
#Extract ONLY U.S. power-sources from dataframe and Remove "United States :" from Power Source column
data_dylan = data_dylan.loc[data_dylan['United States Power Source'].str.contains('United States')]
data_dylan['United States Power Source'] = data_dylan['United States Power Source'].str.lstrip('United States')
data_dylan['United States Power Source'] = data_dylan['United States Power Source'].str.lstrip(': ')
data_dylan['United States Power Source'] = data_dylan['United States Power Source'].str.title()
data_dylan
```
